```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('../dataset/coronavirus-csv/coronavirus_dataset.csv')
```

```python
region = data.groupby(['Country.Region','date'],as_index=False)['cases'].agg('sum')
```

```python
all_unstack = data.groupby(['Country.Region','date'])['cases'].agg('sum').groupby(level=[0]).cumsum().unstack()
```

```python
all_unstack['2020-01-22'] = all_unstack['2020-01-22'].fillna(0)
```

```python
region_cumsum = all_unstack.fillna(method='pad',axis=1).stack().reset_index()
```

```python
region[region['Country.Region']=='Vietnam']
```

```python
region_cumsum[region_cumsum['Country.Region']=='Vietnam']
```

```python
region_cumsum['type']='confirmed'
```

```python
region_cumsum.columns = ['name','date','value','type']
```

```python
region_cumsum = region_cumsum[['name','type','value','date']]
```

```python
region_cumsum
```

```python
region_cumsum.to_csv('../dataset/processed/region_cumsum.csv')
```

```python
region_cumsum[region_cumsum['name']!='Mainland China'].to_csv('../dataset/processed/region_cumsum_no_mainland_china.csv',index=False)
```
